```python
def tkstp(x):
    token=tk.tokenize(x)
    return (' '.join([i for i in token if i.lower() not in stp]))
```

```python
Df.Content=Df.Content.apply(tkstp)
```

* Removing Special  Characters

```python
import re
```

```python
def spcr(x):
    return (re.sub(r'[^a-zA-Z0-9 ]',r'',x))
    
```

```python
Df.Content=Df.Content.apply(spcr)
```

* Lemmatization - keeping root words only

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma=WordNetLemmatizer()
```

```python
def lema(x):
    token=tk.tokenize(x)
    return (' '.join([lemma.lemmatize(i) for i in token]))
```

```python
Df.Content=Df.Content.apply(lema)
```

* Scoring

```python
from afinn import Afinn
```

```python
af=Afinn()
```
